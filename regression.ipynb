```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Algerian_forest_fires.csv", header=1)
```

```python
df.head(10)
```

##### Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

Now the dataset is divided into 2 parts i.e by regions from 0 to 122 we have bejiaia region and after that till end we have Sedi bel abbas region so we will make a new column as region and assign 0 and 1 values 

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fixixng white spaces in the column
df.columns = df.columns.str.strip()
df.columns
```

#### Change the required columns as integer datatype

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != "Classes":
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

### Lets save the cleaned dataset 

```python
df.to_csv("Algerian_forest_fires_cleaned.csv", index=False)
```
